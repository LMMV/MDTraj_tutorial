### Trazar un **mapa de Ramachandran** con matplotlib

```python
from __future__ import print_function
%matplotlib inline
import mdtraj as md
```

#### Cargar trayectoria de la simulación previa

```python
traj = md.load('ala2.h5')
atoms, bonds = traj.topology.to_dataframe()
atoms
```

#### Encontrar los átomos involucrados en los ángulos phi y psi de forma "manual". 

```python
psi_indices, phi_indices = [6, 8, 14, 16], [4, 6, 8, 14]
angles = md.compute_dihedrals(traj, [phi_indices, psi_indices])
```

#### sdfsd

```python
from pylab import *
from math import pi

figure()
title('Dihedral Map: Alanine dipeptide')
scatter(angles[:, 0], angles[:, 1], marker = 'x', c = traj.time)
cbar = colorbar()
cbar.set_label('Time [ps]')
xlabel(r'$\Phi$ Angle [radians]')
xlim(-pi, pi)
ylabel(r'$\Psi$ Angle [radians]')
ylim(-pi, pi)
```
